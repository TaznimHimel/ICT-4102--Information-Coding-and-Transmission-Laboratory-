To develop a program to calculate the channel capacity of a Binary Symmetric Channel (BSC).

## **Objective:**
To write a program to calculate the **capacity of a binary symmetric channel (BSC)**.

## **Theory:**
A **Binary Symmetric Channel (BSC)** is a simple model of a communication channel in which:
- The input is a binary signal (0s and 1s).
- The output is also a binary signal.
- There is a probability of error (denoted as $p$) during the transmission of each bit.
  - The bit is transmitted correctly with probability $1 - p$.
  - The bit is flipped (from 0 to 1 or from 1 to 0) with probability $p$.

The **capacity** of a channel is the maximum rate at which information can be transmitted over that channel with arbitrarily small error. For the BSC, the capacity is given by:

$$
C = 1 - H(p)
$$

Where:
- $C$ is the channel capacity.
- $H(p)$ is the **binary entropy function**, which is defined as:

$$
H(p) = - p \log_2(p) - (1 - p) \log_2(1 - p)
$$

The binary entropy function $H(p)$ quantifies the uncertainty (or entropy) of a random variable with a probability of error $p$.

## **Algorithm:**
1. **Input the probability of error $p$**: This is the likelihood that a transmitted bit will be flipped.
2. **Compute the binary entropy function $H(p)$** using the formula.
3. **Calculate the channel capacity $C$** using $C = 1 - H(p)$.
4. **Display the result**: Show the value of the channel capacity for the given probability $p$.

## **Expected Output:**
1. **Input**: The program will prompt the user to enter the probability of error $p$ (between 0 and 0.5).
2. **Channel Capacity**: The program will compute and display the channel capacity based on the entered value of $p$.

   Example input/output:
   - Enter the probability of error (p) between 0 and 0.5: 0.1
   - Channel Capacity of BSC with error probability 0.1: 0.5310 bits/channel use

3. **Plot**: A plot showing how the **channel capacity** changes as the error probability $p$ increases from 0 to 0.5.

## **Conclusion:**
- The **Binary Symmetric Channel (BSC)** is a simple yet powerful model in communication theory.
- The **channel capacity** increases as the probability of error $p$ decreases.
- When the error probability $p$ is 0, the channel capacity is 1 (maximum capacity).
- As $p$ approaches 0.5, the channel capacity decreases to 0, indicating that the channel is unreliable and unable to transmit information.

```python
import numpy as np
import matplotlib.pyplot as plt

# Binary entropy function
def binary_entropy(p):
    p = np.clip(p, 1e-10, 1 - 1e-10)
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

# Channel capacity function
def bsc_capacity(p):
    return 1 - binary_entropy(p)

# Input
p = float(input("Enter the probability of error (p) between 0 and 1: "))

# Validation
if p < 0 or p > 1:
    raise ValueError("p must be between 0 and 1")

# Calculation
capacity = bsc_capacity(p)

print(f"Channel Capacity of BSC with error probability {p}: {capacity:.4f} bits/channel use")

# Plot
p_values = np.linspace(0, 1, 200)
capacity_values = bsc_capacity(p_values)

plt.plot(p_values, capacity_values, label="Channel Capacity")
plt.title("Binary Symmetric Channel Capacity")
plt.xlabel("Error Probability (p)")
plt.ylabel("Channel Capacity (C)")
plt.grid(True)
plt.legend()
plt.show()
```
