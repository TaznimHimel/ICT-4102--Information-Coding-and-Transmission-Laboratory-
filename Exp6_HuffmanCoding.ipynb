## **Objective:**
To write a program to implement the **Huffman coding algorithm**, which is used for lossless data compression.

## **Theory:**
The **Huffman coding algorithm** is a widely used algorithm for constructing an optimal prefix code based on symbol frequencies or probabilities. Huffman coding is a variable-length prefix code that assigns shorter codes to more frequent symbols and longer codes to less frequent symbols, thereby achieving efficient compression.

The algorithm works by constructing a **binary tree**:
1. **Build a frequency table**: Count the frequency of each symbol.
2. **Build a priority queue (min-heap)**: Insert each symbol into a priority queue with the frequency as the key.
3. **Build the Huffman tree**: Extract the two nodes with the smallest frequencies, combine them into a new node, and insert it back into the queue. Repeat this process until only one node remains.
4. **Assign binary codes**: Assign binary digits (0 and 1) to the tree's branches, where the left branch is assigned 0 and the right branch is assigned 1. The codes for the symbols are determined by the path from the root to the symbol's leaf node.

The **Huffman code** is optimal in the sense that it minimizes the average code length, which in turn minimizes the total number of bits needed to encode a message.

## **Algorithm:**
1. **Input the symbols and their frequencies**.
2. **Build a frequency table**.
3. **Build a priority queue (min-heap)** based on the frequencies.
4. **Build the Huffman tree** by repeatedly combining the two least frequent nodes.
5. **Assign binary codes** to the symbols.
6. **Output the Huffman codes** for each symbol.

## **Conclusion:**
- The **Huffman coding algorithm** efficiently compresses data by assigning shorter codes to more probable symbols and longer codes to less probable symbols.
- The algorithm works by constructing a **binary tree** where the path from the root to each symbol's leaf node determines the binary code for that symbol.
- This is a **lossless compression algorithm**, meaning the original data can be recovered from the compressed data without any loss.

## **Code:**


```python
import numpy as np

# ---- Input ----
symbols = ['x1','x3','x5','x2','x4']
prob = np.array([0.4, 0.2, 0.2, 0.15, 0.05])

# ---- Normalize ----
prob = prob / np.sum(prob)

# ---- Initialize nodes ----
nodes = []
for i in range(len(symbols)):
    nodes.append({
        "symbols": [symbols[i]],
        "prob": prob[i]
    })

# ---- Code storage ----
codes = {s: "" for s in symbols}

# ---- Huffman Tree Construction ----
while len(nodes) > 1:

    # Sort nodes by probability (ascending)
    nodes = sorted(nodes, key=lambda x: x["prob"])

    left = nodes[0]
    right = nodes[1]

    # Assign 0 and 1
    for s in left["symbols"]:
        codes[s] = '0' + codes[s]

    for s in right["symbols"]:
        codes[s] = '1' + codes[s]

    # Merge nodes
    new_node = {
        "symbols": left["symbols"] + right["symbols"],
        "prob": left["prob"] + right["prob"]
    }

    # Remove two smallest and add merged node
    nodes = nodes[2:]
    nodes.append(new_node)

# ---- Results ----
avg_length = 0
entropy = 0

print("Symbol   Prob     Code    Length")
print("--------------------------------")

for i in range(len(symbols)):
    code = codes[symbols[i]]
    L = len(code)

    avg_length += prob[i] * L
    entropy -= prob[i] * np.log2(prob[i])

    print(f"{symbols[i]:<8}{prob[i]:<8.3f}{code:<8}{L}")

# ---- Final Metrics ----
print(f"\nAverage Length = {avg_length:.3f} bits")
print(f"Entropy = {entropy:.3f} bits")
print(f"Efficiency = {(entropy/avg_length)*100:.2f} %")
```
